## 디지털전환(Digital Transformation)

- **Digitization:** `아날로그` 데이터의 `디지털화`

- **Digitalization:**
> - (좁은의미) `데이터 유형의 디지털화`를 포함한 `프로세스의 디지털화`, 간단히 **`디지털화`** 라고 함
> - (넓은의미) 똑똑해지는 전자기기들로 `모든 정보들이 컴퓨터 기반 디지털 형식으로 변환`되고 효율성과 생산성을 높이기 위해 `자동화 되는 프로세스`

- **Digital Transformation:** `프로세스의 디지털화`를 포함한 `비즈니스 산업 전반의 변화`까지 포함하는 것으로, 근본적으로 `새로운 방법을 모색하거나 재구상하는 활동 전반`을 의미

## 디지털경제(Digital Economy)

> **"`디지털화를 기반`으로 `데이터가 생성`되고 `연결`되는 사회경제적 표현"**
- **Open API(Open Application Programming Interface):** 컴퓨터 프로그래밍으로 구현되어 `데이터를 공개 및 공유`하여 누구나 사용할 수 있도록 `권한을 제공`하는 것
- **데이터기반(Data Driven):** `데이터를 활용 또는 주도`하여 무언가를 하는 `활동 자체`를 의미
- **데이터비즈니스(Data Business):** 데이터를 활용 또는 주도하여 `비즈니스를 개선`하는 것을 의미하며 **`디지털 비즈니스`** 또는 **`플랫폼 비즈니스`** 라고도 함

## 플랫폼경제(Platform Economy)

> **"`디지털 경제 ∋ 플랫폼 경제`"**

**"`빅데이터`를 다루고 `자동으로 의사결정`하는 `인공지능` 기술까지 `자동화한 플랫폼`을 사용 `디지털경제` 구축"**

- **Platform:** `초기`에는 `빅데이터를 수집`하는 **`정보시장`** 을 의미하였으나, 점차 인공지능 기술까지 더해져 `자동으로 인간을 대신하여 의사결정`을 해주는 기능까지로 확장
> - `디지털 전환`이 가속화되고, `인공지능` 기술이 발달함에 따라 `플랫폼 경제`가 큰 화두
> - **대표적 플랫폼 기업:** `유튜브, 구글, 아마존, 페이스북, 알리바바, 에어비앤비, 네이버, 카카오 등`

# **디지털생태계를 위한 데이터싸이언스 시대**

## 빅데이터? 인공지능? 데이터싸이언스?

- 디지털전환에 의해 `플랫폼 경제 또는 디지털 경제`를 사용하는 `디지털생태계를 완성`시켜줄 **`3대 핵심 기술`**
- 3가지 핵심 기술 중 `하나라도 없으면` 어떠한 디지털 비즈니스나 산업, 경제도 `완성되지 않음`
- 디지털생태계까지는 사회적, 기술적 방향만을 의미하며 `3대 핵심 기술`은 **`현실 문제 해결`**에 집중된 용어
> - **빅데이터(BigData):** 디지털전환과 연결로 생성되는 `수많은 데이터`
> - **인공지능(Artificial Intelligence, AI):** 빅데이터에서 `자동으로 의사결정`하는 알고리즘
> - **데이터싸이언스(Data Science) = 데이터분석(Data Analysis):** 비즈니스 용어로 빅데이터와 인공지능을 활용하여 `실제 문제를 해결하는 과정 전체` 의미

> **1) 1차 산업혁명(약 100여년): 농사 짓다 보니 공장 노동자가 됐다**
> - 산업혁명 어원은 영국의 역사학자이자 문명비평가인 Arnold Joseph Toynbee(1889~1975)가 처음 사용
> - 영국에서 시작해 전세계로 전파되었고 글로벌화와 함께 역사학자가 만든 용어는 경제학 분야의 학술 용어로 현재 더욱 알려짐
> - 증기를 이용한 인류 역사상 **`최초의 근대공장으로 자동화 도입`**
> - 식량을 스스로 **`생산 하는 시대`** 에서 **`구매하는 시대`**로 바뀌며 새로운 소비층과 도시가 발달
> - **`임금을 받는 노동자 계급`** 이 생기며 농장 근로자가 공장 근로자로 전환

> **2) 2차 산업혁명(약 70~80년): 새로운 동력원의 등장과 발전**
> - 석탄을 사용 대신 **`석유와 전기`** 라는 새로운 동력 등장
> - 새로운 동력은 **`강가 주변으로 공장을 이동`** 시킴
> - 동력의 부피가 작아지고 공장의 생산성이 높아지며 **`자동차, 식료품, 음료, 영화, 라디오, 축음기 등`** 다양하게 보급

> **3) 3차 산업혁명(약 50~60년): 컴퓨터와 인터넷으로 경제 다양화**
> - 사람들은 **`컴퓨터와 인터넷을 결합`** 해 커뮤니케이션과 경제에 집중 사용
> - 공유가 수월해지면서 **`수직적 권력이 수평적으로 전환`**
> - 향후 등장하는 4차 산업혁명도 3차 산업혁명의 과정 중 하나라고 주장하기도

> **4) 4차 산업혁명:**
> - 독일 경제학자 Klaus Schwab가 세계 경제 포험에서 주창한 용어
> - **`빅데이터, 인공지능, 로봇공학, 사물인터넷, 무인항공기, 3차원인쇄, 나노기술 등`** 6대 기술혁신 토대
> - 기존 바보였던 IT제품들이 특정 **`명령(Rule)`** 을 수행했다면, 스스로 **`학습(Training)`** 하여 주어진 환경에서 최상의 결과 도출
> - 1차 산업혁명은 무려 백여년 동안 이어졌으나 4차 산업혁명은 체감조차 되지 않는 신생아 수준
> - **`모든 것이 연결되고 보다 지능적인 사회`** 로 프레임이 진화되기 시작

## 왜 데이터싸이어티스트가 필요한가?

**1) 디지털 경제/비즈니스에서 새로운 것은 무엇인가?**
> - 현명한 비즈니스 의사결정을 위해 기업은 경영진 + 조직 + <span style="color:red">**기술**</span>을 끊임없이 변화
>> - 비즈니스 기업들이 **`IT(Information Technology)`** 와 **`SI(System Integration)`** 에 투자하는 이유는, 실질적인 경제가치를 제공
>> - 생산성의 증가, 수익의 증가, 기업가치(주식)의 증가, 장기적 미래수익 보장
>> - 더 나은 경영 의사결정, 더 효율적인 비즈니스 프로세스, 더 높은 기업 수익성
>
> <center><img src='Image/Basic/Programming_GlobalTrend.png' width='600'>(다보스 포럼, 2016)</center>
>
> **(1) IT혁신**
> - 스마트폰과 태블릿, 클라우드 컴퓨팅을 기반으로 한 모바일 디지털 비즈니스 플랫폼의 등장
> - 새로운 제품과 서비스가 창출되며 기존 사업이 파괴되고 새로운 비즈니스 모델 개발 박차
>
> **(2) 전자상거래 확장**
> - 페이스북, 유튜브, 트위터, 텀블러, 넷플릭스, 애플 뮤직 등은 새로운 미디어 플랫폼을 제시
> - 물리적인 제품 판매에서 서비스 판매라는 새로운 가치 흐름 모델 제시
>
> **(3) 새로운 비즈니스 모델**
> - 전통적인 마케팅과 광고산업을 교란시키고 주요 미디어와 콘텐츠 공급자들을 위험에 빠뜨림
> - 넷플릭스 서비스는 미디어 배포/생성 방식을 완벽하게 바꾸어, 미디어 및 방송 제작자와 매니지먼트 네트워크 지배력을 새롭게 창조
> - 아이튠즈는 TV 쇼 다운로드의 67%를 차지하고 시청자들을 TV에서 인터넷 환경으로 옮겨놓음
>
> <span style="color:blue">**(4) 기업 운영방식과 조직의 변화**</span>
> - 기업 조직의 상위에서 하위까지의 커뮤니케이션과 감독방식이 훨씬 빠르고 가까워짐
> - 풍부한 데이터 기반 정보시스템으로 정보가 빠르게 공유되고 즉각적인 문제 이해와 의사결정 가능
> - 기존 직원들이 단편적인 업무를 담당하였던 반면, 여러가지 역할과 업무를 동시에 해결 및 협업 강조
> - 서열상 역할과 위치보다 역량과 기술, 그리고 데이터 분석에 기초한 빠르고 정확한 의사결정 요구
> - 소비자의 기술, 태도, 문화를 빠르게 파악하고 접근할 수 있는 방향으로 모든 방식이 변화 중
>> - 조직의 가치, 구조 및 행동 양식과 기타 보완적 변화가 동반하지 못한다면 투자는 비용으로 전락
>> - 실질적인 새로운 디지털플랫폼의 혜택을 위해 업무 수행 방식과 문화 변경이 필수적
>> - 새로운 비즈니스 모델, 새로운 프로세스, 새로운 경영 행위, 새로운 조직 문화와 훈련 등이 성과 극대화 (Brynjolfsson, 2005; Brynjolfsson and Hitt, 2000; Laudon, 1974)
>>

**2) 디지털 전환으로 모든 사람들에게 도전과 기회가 수평화 되었는가?**
- **콜럼버스의 신대륙 발견:** 세계 무역이 발달되고 글로벌 사람들의 교류를 증대 시켜 문화를 가깝게 만듬
- **산업혁명:** 국가 간의 무역을 확대시키고 글로벌 경제를 최초로 출현하게 함
- **인터넷과 글로벌 통신:** 선진국과 후진국의 경제적 문화적 이점이 빠르게 공유
> **"하지만.."**
> - 선진국의 생존과 일자리 창출, 시장과 자원, 그리고 아이디어에서 저개발국가의 저임금지역에서 고등교육을 받고 동기부여가 충실한 사람들과 끊임없는 전쟁 중(Freidman, 2007)
> - 전통적인 국가의 경계가 약화되고 비즈니스 기업들에게 도전과 기회를 마련
> - 저임금 생산자로 인해 미국은 지난 10년간 700만개의 제조업 일자리 사라짐
> - 제조업 비중은 줄어들고, 건축설계/금융/컨설팅/공학/의학 등 **`시장성 있는 서비스`** 비중 증가
> - IT혁명으로 운영 및 거래비용은 극적으로 감소됨
> - 구글 이베이와 같은 서비스 기업은 고정비용이 높은 정보시스템을 재설계하지 않고 저임금 글로벌 복제
- **경제적 파급:** 거래비용과 대리인비용을 모두 줄여 디지털전환에 따른 기업 규모의 축소 예상
> - **거래비용 이론(Transaction Cost Theory):** 기업/개인은 생산비용을 절약하려고 노력하는 것처럼 거래비용을 절약하려 함
>> - 기업이 직접 제품/서비스를 생산하는 것보다 재화/용역을 구매가 더 쉽고 저렴하기에 기업 규모(직원 수)가 줄어듬
>> - 직원 고용보다 경쟁시장에 외주하는 것이 훨씬 비용이 저렴하기 때문에 기업들은 규모(직원 수)를 줄일 수 있음
> - **대리인 이론(Transaction Cost Theory):** 사장(소유자)은 그(그녀)를 대신하여 일할 "대리인"(직원)을 고용하지만, 대리인을 관리/감독 할 수 없으면 대리인은 소유자의 이익보다 자신의 이익을 추구하려는 경향
>> - 회사의 규모/범위가 커지면 직원들을 관리/감독하기 위해 더 많은 노력이 필요해서 대리인비용 또는 조정비용이 증가
>> - 디지털 전환은 정보 획득 및 분석비용을 줄여 관리자들이 많은 직원들을 쉽게 관리할 수 있게 되기에 대리인비용을 절감
>> - 전반적인 관리비용이 줄면 기업은 중간 관리자와 사무직의 수를 줄이면서 수익을 늘릴 수 있음

**3) 우리가 가야할 방향은?**
> **(1) 디지털 기업으로의 전환(Digital Transformation)**
>
> - 고객/공급업체/종업원 사이의 중요한 **`비즈니스 관계`** 는 디지털 방식으로 구성
> - 핵심 **`비즈니스 프로세스`** 는 기업 전반과 조직 간 디지털 네트워크를 통해 수행
>> - 신제품 개발, 주문처리, 마케팅 계획수립, 채용 등은 비즈니스 프로세스의 예
> - **`핵심 기업 자산`** 은 디지털 방식으로 관리
>> - 지적재산, 핵심역량, 금융 및 인적자원
>> - 관련된 의사결정 필요 정보들을 언제 어디서나 이용할 수 있는 공유 플랫폼 환경
> - 전통적인 기업보다 환경변화를 신속하게 감지 및 대응하여 **`생존을 위한 유연성 확보`**
>> - **시간변경:** 특정 노동시간이 아닌 24시간 지속적으로 수행되는 비즈니스
>> - **공간변경:** 특정 위치가 아닌 범세계적 작업공간이 활용되는 비즈니스
>> - 20세기 비즈니스 방식이 **`사람`** 이었다면 21세기 비즈니스 방식은 **`정보기술`**
>
> **(2) 디지털플랫폼의 전략적 비즈니스 활용**
>> **[1] 저비용 리더십:** 운영비용을 줄이고 가격을 낮추기 위해 디지털 플랫폼을 이용하라!(월마트)
>>
>> **[2] 제품 차별화:** 새로운 제품/서비스를 만들거나 기존 제품/서비스 향상을 위해 디지털 플랫폼을 이용하라!(우버, 나이키, 애플)
>>
>> **[3] 틈새시장 집중:** 특정 시장에 집중하고 좁은 목표시장에서 경쟁자 우위를 위해 디지털 플랫폼을 이용하라!(힐튼 호텔, 하라스)
>>
>> **[4] 고객과 공급업자와의 유대 강화:** 고객과 공급업자와의 유대 강화를 위해 디지털 플랫폼을 이용하라!(토요타, 아마존)
> - 디지털 플랫폼은 전략적 비즈니스 목표 달성 뿐 아니라 미국과 선진국에서 `일상적인 비즈니스 수행 도구`
>> - 여행, 의료, 교육, 개인서비스, 금융, 보험, 부동산 등 정보시스템 없는 운영은 불가능
> - 기업의 목표달성과 정보기술 활용도의 상호 의존성 증가
>> - 미래의 비즈니스는 현재의 정보시스템이 어떤 일을 할 수 있는가에 달림
>> - 운영 최적화, 새로운 제품 개발, 서비스 비즈니스 모델링, 고객과 공급자 친밀성, 의사결정 환경개선, 경쟁우위 실시간 파악, 생존 확률 계산 모델링
>> - `디지털자산들의 관계를 이해/활용`하고 `빠르게 변화`시킬수록 `유능한 경영자`

> **(3) 사회적, 조직적, 지식적 융합 필요**
> - **OR(Operations Research):** 수송, 재고관리, 거래비용 등의 조직 내 매개변수들을 최적화하는 수리적 기법
> - **컴퓨터과학:** 계산가능성, 계산방법론, 효율적 데이터저장, 데이터 전처리 등의 개발 관련 기법
> - **경영과학:** 의사결정과 경영 업무를 위한 모형(방법론, 모델링) 개발 중점
> - **사회학:** 정보시스템이 개인, 집단, 조직에 어떤 영향을 미치는지 연구nn
> - **경제학:** 디지털 제품 생산의 이해, 시장의 변화, 정보시스템이 기업 내 비용 구조에 미치는 영향 연구
> - **심리학:** 의사결정자가 정보시스템을 통해 어떻게 정보를 인지하고 사용하는지 연구
